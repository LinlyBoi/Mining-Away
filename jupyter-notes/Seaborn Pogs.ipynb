Below is a simple *Scatter plot* with histograms and splines.

```python
import seaborn as sns

# Loading a dataset that comes with Seaborn
tips_dat = sns.load_dataset("tips")

tip_plot = sns.JointGrid(x = "total_bill", y = "tip", data = tips_dat);
tip_plot.plot_joint(sns.histplot)
tip_plot.plot_marginals(sns.kdeplot)

sns.relplot(
    data=tips_dat,
    x="total_bill", y="tip", hue="smoker", col="time",
)
```

```python
flights_dat = sns.load_dataset("flights")

sns.relplot(data=flights_dat, x="year", y="passengers", hue="month", kind="line")
sns.relplot(data=flights_dat, x="month", y="passengers", hue="year", kind="line")
```

```python
anscombe_dat = sns.load_dataset("anscombe")

sns.lmplot(
    data = anscombe_dat, x = "x", y = "y", col = "dataset", hue = "dataset",
    col_wrap = 2, palette = "muted", ci = None,
    height = 4, scatter_kws = {"s": 50, "alpha": 1}
)
```
